## create Engine SqlAlchemy

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///chinook.sqlite')
```

```python
engine.table_names()
```

### select all firstname , lastname ,customerid,country not  in 'us'

```python
query = '''
    SELECT firstname, lastname, customerid, country
    from Customer
    where country != 'USA'
'''
```

```python
pd.read_sql_query(query,engine)
```

## select customers from brazil

```python
pd.read_sql_query("SELECT * from Customer WHERE country = 'Brazil'",engine , index_col='CustomerId')
```

### provide a query the invoices of customers who are from brazil. the resturant table should show the customer's full name , invoice id ,data of the invoice and billing country

```python
pd.read_sql_query("""SELECT c.firstname,c.lastname,i.invoiceid,i.InvoiceDate
from Customer as c , Invoice as i 
WHERE country = 'Brazil' and c.customerid = i.CustomerId""",engine)
```

```python
query = """
    SELECT e.firstname , e.lastname,c.firstname,c.lastname , i.total
    from Employee as e
    join Customer as c on c.customerid = i.customerid
    JOIN Invoice as i on c.customerid = i.CustomerId
"""
pd.read_sql(query,engine)
```

## how many invoices between 2009 and 2011

```python
query = """
    SELECT count(*)
    from Invoice 
    where invoicedate BETWEEN datetime('2009-01-01 00:00:00') and datetime('2011-01-01 00:00:00')
"""
pd.read_sql(query,engine)
```
